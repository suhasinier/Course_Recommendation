```python
# Importing all necessary libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the Excel file

```python
courses = pd.read_excel('course_recommendation.xlsx')
```

```python
courses.head()
```

#### Description of the data
## The file contains 4 features about the course which have been uploaded into our database
* title - Title of the Course
* technology - Technology of the course 
* topic - course topic
* abstract - Short summary of the course
* url - Course url - if applicable

```python
# Fill the missing values with empty string
```

```python
courses.fillna(" ",inplace=True)
```

```python
courses.head()
```

```python
courses.shape
```

```python
ratings=pd.read_excel("rating.xlsx")
ratings.head()
```

```python
# Filterig users who have rated more than 3 courses
x=ratings['userid'].value_counts() >3
x[x].shape
```

```python
# Getting indices of users who have rated more than 3 courses
y=x[x].index


# Filterig rating dataframe for users who have rated more than 3 courses

ratings=ratings[ratings['userid'].isin(y)]
ratings.head()
```

```python
# Merging ratings and coursees dataframe on userid

courses_with_rating = courses.merge(ratings,on='courseid')
courses_with_rating.head()
```

```python
courses_with_rating.shape
```

```python
# Counting the number of ratings wrt title

num_rating = courses_with_rating.groupby('title')['rating'].count().reset_index()
num_rating.head()
```

```python
num_rating.shape
```

```python
# Rename rating as numbe rof rating
num_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
num_rating.head()
```

```python
# Merging num_rating with sourses_with_rating on title 
courses_with_rating = courses_with_rating.merge(num_rating,on='title')
courses_with_rating.head()
```

```python
# Creating a final dataframe where number of ratings >=2
final_rating = courses_with_rating[courses_with_rating['rating']>=2]
final_rating.sample(10)
final_rating.shape
```

```python
final_rating.drop_duplicates(['userid','title'],inplace=True)
final_rating.head()
final_rating.shape
```

```python
final_rating.head()
```

```python
# creatign a pivot table to find 
courses_pivot = final_rating.pivot(columns='userid',index='title',values='rating')
courses_pivot.head()
```

```python
courses_pivot.fillna(0,inplace=True)
courses_pivot.head()
```

### To learn about Sparse Matrix
https://www.learndatasci.com/glossary/sparse-matrix/

```python
from scipy.sparse import csr_matrix
courses_sparse = csr_matrix(courses_pivot)
```

```python
print(courses_sparse)
```

```python
print(f"Indices : {courses_sparse.indices}\n")
print(f"Values  : {courses_sparse.data}\n")
print(f"Idnex Pointer  : {courses_sparse.indptr}\n")
```

```python
# Model Building

from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')


# Fitting the model
model.fit(courses_sparse)
```

```python
# Finding the distance and suggestion for a course with random index value

distance,suggestion = model.kneighbors(courses_pivot.iloc[15,:].values.reshape(1,-1),n_neighbors=6)
```

```python
# Distance of all six suggested courses  with course having index 15
print(distance)
print("\n")

# Course indices of all six suggested courses  
print(suggestion)

print("\n")
# Finding the title of suggested courses

for i in range(len(suggestion)):
    print(courses_pivot.index[suggestion[i]])
```

```python
#  Storing the course names for creating the page  (for pickl file)

course_name = courses_pivot.index
course_name
```

```python
import pickle
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(course_name,open('artifacts/course_name.pkl','wb'))
pickle.dump(courses_pivot,open('artifacts/courses_pivot.pkl','wb'))
pickle.dump(final_rating,open('artifacts/final_rating.pkl','wb'))
```
