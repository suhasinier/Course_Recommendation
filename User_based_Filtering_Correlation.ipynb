```python
# Importing all necessary libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the Excel file

```python
courses = pd.read_excel('course_recommendation.xlsx')
```

```python
courses.head()
```

#### Description of the data
## The file contains 4 features about the course which have been uploaded into our database
* title - Title of the Course
* technology - Technology of the course 
* topic - course topic
* abstract - Short summary of the course
* url - Course url - if applicable

```python
# Fill the missing values with empty string
```

```python
courses.fillna(" ",inplace=True)
```

```python
courses.head()
```

```python
ratings=pd.read_excel("rating.xlsx")
ratings['rating'] = ratings['rating'].round(1)
ratings.head()
```

```python
ratings['userid'].value_counts().shape
```

```python
# Filterig users who have rated more than 3 courses
x=ratings['userid'].value_counts() >3
x[x].shape
```

```python
# Getting indices of users who have rated more than 3 courses
y=x[x].index


# Filterig rating dataframe for users who have rated more than 3 courses

ratings=ratings[ratings['userid'].isin(y)]
ratings.head()
```

```python
# Merging ratings and coursees dataframe on userid

courses_with_rating = courses.merge(ratings,on='courseid')
courses_with_rating.head()
```

```python
courses_with_rating.shape
```

```python
number_rating = courses_with_rating.groupby('title')['rating'].count().reset_index()
number_rating
```

```python
number_rating.rename(columns={'rating':'number_rating'},inplace=True)
number_rating.head()
```

```python
courses_with_rating.head()
```

```python
# Merging num_rating with courses_with_rating on title 

courses_with_rating = courses_with_rating.merge(number_rating,on='title')
courses_with_rating.head()
```

```python
courses_with_rating.shape
```

```python
# Creatign a final dataframe with rating > 3

final_rating = courses_with_rating[courses_with_rating['rating']>=2]

print(final_rating.shape)
final_rating.head()
```

```python
final_rating.drop_duplicates(['userid','title'],inplace=True)
print(final_rating.shape)
final_rating.head()
```

```python
# creatign a pivot table to find 
courses_pivot = final_rating.pivot(index='userid',columns='title',values='rating')
courses_pivot.head()
```

```python
courses_pivot.fillna(0,inplace=True)
courses_pivot.head()
```

```python
# Input a course name to recommend
#course_name = input("Enter the course name for recommendation")
```

```python

def recommend_course(course_name):
    user_ratings_for_course = courses_pivot[course_name]
    similar_course = courses_pivot.corrwith(user_ratings_for_course)
    corr_course_name = pd.DataFrame(similar_course,columns=['Correlation'])
    corr_course_name.dropna(inplace=True)
    courses_recommended = corr_course_name.sort_values('Correlation',ascending=False).head(5)
    print(courses_recommended)
```

```python
recommend_course('3-Axis Machining with Autodesk Fusion 360')
```

```python
recommend_course('ACT Prepaaration')
```
